```python
import pandas as pd
df = pd.read_csv('test.csv')
df
```

```python
df = df.drop(columns =['PassengerId','Pclass','Name','Sex','SibSp','Parch','Ticket','Cabin','Embarked'])
df
```

```python
df['Age']. fillna(df.Age.median(),inplace=True)
```

```python
df.isnull (). sum()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop('Survived', axis=1), df['Survived'], test_size =0.3,random_state =0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train["Fare"], X_train["Age"])
plt.scatter(X_test["Fare"], X_test["Age"])
plt.show()
```

```python
import matplotlib.pyplot as plt
sorted_data = X_train.sort_values(by = 'Age')
sorted_data2 = X_test.sort_values(by = 'Age')
plt.plot(sorted_data["Age"], sorted_data["Fare"])
plt.plot(sorted_data2["Age"], sorted_data2["Fare"])
plt.show()
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler ()
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train ); X_test_norm = scaler.transform(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter([j for i, j in X_train_norm], [i for i, j in X_train_norm])
plt.scatter([j for i, j in X_test_norm], [i for i, j in X_test_norm])
plt.show()
```

```python
import matplotlib.pyplot as plt
arr = {i:j for i, j in X_train_norm}
arr = sorted(arr.items())
plt.plot([i for i, j in arr], [j for i, j in arr])
plt.show()
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler2 = StandardScaler ()
scaler2.fit(X_train)
X_train_stand =scaler2.transform(X_train ); X_test_stand = scaler2.transform(X_test)
```
